```python
import pandas as pd
import numpy as np
from sklearn.metrics import accuracy_score
```

```python
credit_card_data=pd.read_csv('creditcard.csv')
```

```python
credit_card_data
```

```python
credit_card_data.info()
```

```python
credit_card_data.isnull().sum()
```

```python
credit_card_data['Class'].value_counts()
```

```python
legil=credit_card_data[credit_card_data.Class==0]
fruad=credit_card_data[credit_card_data.Class==1]
```

```python
print(legil.shape)
print(fruad.shape)
```

```python
new_legit=legil.sample(n=492)
```

```python
new_data=pd.concat([new_legit,fruad],axis=0)
```

```python
new_data.head()
```

```python
new_data.value_counts('Class')
```

```python
new_data['Class'].value_counts()
```

```python
new_data.groupby('Class').mean()
```

```python
X=new_data.drop(['Class'],axis=1)
y=new_data['Class']
```

```python
print(X)
print(y)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,stratify=y,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
y_test
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
```

```python
model.fit(X_train,y_train)
```

```python
from sklearn.metrics import accuracy_score
```

Accuracy on training data

```python
X_train_prediction=model.predict(X_train)
training_data_accuracy=accuracy_score(X_train_prediction,y_train)
print(f'The accuracy score on traning data : {training_data_accuracy}')
```

Accuracy on testing data

```python
X_test_prediction=model.predict(X_test)
testing_data_accuracy=accuracy_score(X_test_prediction,y_test)
print(f"The Accuracy Score on Testing Data : {testing_data_accuracy} ")
```

```python
model.score(X_test,y_test)
```
